#### Location Delhi

```python
# Scraping name
name_delhi = driver.find_elements_by_xpath('//h2[@class="_3FpCg"]')
for i in name_delhi:
    name.append(i.text)

# Scraping model
model_delhi = driver.find_elements_by_xpath('//p[@class="cvakB"]')
for i in model_delhi:
    model.append(i.text)

# Scraping km_driven
km_driven_delhi = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[1]')
for i in km_driven_delhi:
    km_driven.append(i.text)

# Scraping owner
owner_delhi = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[2]')
for i in owner_delhi:
    owner.append(i.text)
    
# Scraping fuel type
fuel_type_delhi = driver.find_elements_by_xpath('//ul[@class="bVR0c"]/li[3]')
for i in fuel_type_delhi:
    fuel_type.append(i.text)
    
# Scraping price
price_delhi = driver.find_elements_by_xpath('//div[@class="_7udZZ"]/span')
for i in price_delhi:
    price.append(i.text)
```
