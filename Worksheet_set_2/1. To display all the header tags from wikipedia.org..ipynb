### Display all the header tags from wikipedia.org. 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = requests.get(r'https://en.wikipedia.org/wiki/Main_Page')
url
```

```python
wiki = BeautifulSoup(url.content)
wiki
```

```python
title = wiki.find_all(['h1','h2','h3','h4','h5','h6'])
print('List of all header tags from wikipedia.org are :',*title,sep='\n\n')
```
