### Top 10 ODI teams in women’s cricket along with the records for matches, points and rating from icc-cricket.com

```python
import numpy as np
import pandas as pd
import requests as rq
import warnings

from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.icc-cricket.com/rankings/womens/team-rankings/odi')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
col = []
for i in Data.find_all('th'):
    col.append(i.text.replace('\n',''))
col    
```

```python
Table_Data = Data.find_all('td')

Table_Data
```

```python
Table_List = []
start = 0
end = 5
for i in range(10):
    Position,Team,Matches,Point,Rating = Table_Data[start:end]
    Table_List.append([Position.text.replace('\n',''),Team.text.replace('\n',''),Matches.text.replace('\n',''),Point.text.replace('\n',''),Rating.text.replace('\n','')])
    start+= 5
    end +=5
    
```

```python
Table_List
```

```python
df = pd.DataFrame(Table_List,columns=col)
df
```

```python
df['RatingR'] = df['RatingR'].str.strip()
df
```

```python
df.rename(columns={'Pos':'Rank','TeamT':'Team','MatchesM':'Matches','PointsP':'Points','RatingR':'Rating'},inplace=True)
df
```

```python
df['Points'] = df['Points'].str.replace(',','')
df
```

```python
df.info()
```

```python
df = df.astype({'Rank':int,'Rating':int})
df.info()
```

```python
date = Data.find('div',class_='rankings-table__last-updated').text
date
```

```python
df.style.set_caption(date)
```

```python
df.to_csv('5(a). Top 10 ODI teams in women’s cricket.csv')
```
