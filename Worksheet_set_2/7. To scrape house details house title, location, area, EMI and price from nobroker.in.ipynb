### To scrape house details house title, location, area, EMI and price from nobroker.in

```python
import numpy as np
import pandas as pd
import requests as rq
import warnings as wr

from bs4 import BeautifulSoup
wr.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.nobroker.in/property/sale/bangalore/Electronic%20City?type=BHK4&searchParam=W3sibGF0IjoxMi44N%20DUyMTQ1LCJsb24iOjc3LjY2MDE2OTUsInBsYWNlSWQiOiJDaElKdy1GUWQ0cHNyanNSSGZkYXpnXzhYRW8%20iLCJwbGFjZU5hbWUiOiJFbGVjdHJvbmljIENpdHkifV0=&propertyAge=0&radius=2.0')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
House_Title = []
for i in Data.find_all('h2',class_='heading-6 font-semi-bold nb__25Cl7'):
    House_Title.append(i.text)
House_Title    
```

```python
Location = []
for i in Data.find_all('div',class_='nb__1EwQz'):
    Location.append(i.text)
Location    
```

```python
Area = []
for i in Data.find_all('div',attrs={'class':'font-semi-bold heading-6','id':'minRent'}):
    Area.append(i.text)
Area    
```

```python
Emi = []
for i in Data.find_all('div',attrs={'class':'font-semi-bold heading-6','id':'roomType'}):
    Emi.append(i.text)
Emi    
```

```python
For_Price = Data.find_all('div',attrs={'class':'nb__4L90a','id':'minDeposit'})
For_Price
```

```python
Price = []
for i in For_Price:
    Price.append(i.find_all('span')[0].text)
    
Price    
```

```python
Unit = []
for i in For_Price:
    Unit.append(i.find_all('div',class_='heading-7')[0].text)
    
Unit
```

```python
Loca = Data.find_all('span',itemprop='geo')
Loca
```

```python
Latitude = []
for i in Loca:
    Latitude.append(i.find_all('meta',itemprop='latitude')[0]['content'])
Latitude    
```

```python
Longitude = []
for i in Loca:
    Longitude.append(i.find_all('meta',itemprop='longitude')[0]['content'])
Longitude    
```

```python
df = pd.DataFrame({'House_Title':House_Title,'Address':Location,'Latitude':Latitude,'Longitude':Longitude,'Area':Area,'EMI':Emi,'Price':Price,'Price Per Unit':Unit})
df
```

```python
df.to_csv('7. To scrape house details house title, location, area, EMI and price from nobroker.in.csv')
```
