### Display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) using data frame 

```python
from bs4 import BeautifulSoup
import requests as rq
import pandas as pd
```

```python
url_1_to_50 = rq.get(r'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
url_1_to_50
```

```python
url_51_to_100 = rq.get(r'https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
url_51_to_100
```

```python
Data_1_to_50 = BeautifulSoup(url_1_to_50.content)
Data_1_to_50
```

```python
Data_51_to_100 = BeautifulSoup(url_51_to_100.content)
Data_51_to_100
```

```python
movies_1_to_50 = []

for i in Data_1_to_50.find_all('h3',class_='lister-item-header'):
    movies_1_to_50.append(i.text)
movies_1_to_50    
```

```python
movie_1_to_50 = []

for i in Data_1_to_50.find_all('h3',class_='lister-item-header'):
    movie_1_to_50.append(i.text.replace('\n',' '))
movie_1_to_50 
```

```python
movies_51_to_100 = []
for i in Data_51_to_100.find_all('h3',class_='lister-item-header'):
    movies_51_to_100.append(i.text)
movies_51_to_100    
```

```python
movie_51_to_100 = []
for i in Data_51_to_100.find_all('h3',class_='lister-item-header'):
    movie_51_to_100.append(i.text.replace('\n',' '))
movie_51_to_100 
```

```python
df1 = pd.DataFrame({'Movies':movie_1_to_50})
df1
```

```python
df1['Rank'] = df1['Movies'].str.split('.').str.get(0)
df1
```

```python
df1['Name Of Movie'] = df1['Movies'].str.split('(',n=1).str.get(0).str.split('.',n=1).str.get(1)
df1
```

```python
df1['Releasing Year'] = df1['Movies'].str.split('(',n=1).str.get(1).str.replace(')',' ')
df1
```

```python
df1 = df1.drop(columns='Movies')
df1
```

```python
Length_1_to_50 = []

for i in Data_1_to_50('span',class_='runtime'):
    Length_1_to_50.append(i.text)
Length_1_to_50    
    
```

```python
df1['Duration'] = Length_1_to_50
df1
```

```python
type_1_to_50 = []

for i in Data_1_to_50('span',class_='genre'):
    type_1_to_50.append(i.text.replace('\n',' '))
type_1_to_50    
```

```python
df1['Type Of Movie'] = type_1_to_50
df1
```

```python
rating_1_to_50 = []
for i in Data_1_to_50('div',class_='inline-block ratings-imdb-rating'):
    rating_1_to_50.append(i.text.replace('\n',' '))
rating_1_to_50    
```

```python
df1['Rating'] = rating_1_to_50
df1
```

```python
Votes_1_to_50 = []
for i in Data_1_to_50('p',class_='sort-num_votes-visible'):
         Votes_1_to_50.append(i.text.replace('\n',' ').split('|')[0].replace('Votes:',' '))
Votes_1_to_50        
    
```

```python
print(len(Votes_1_to_50))
```

```python
gross_1_to_50 = []
for i in Data_1_to_50('p',class_='sort-num_votes-visible'):
         gross_1_to_50.append(i.text.replace('\n',' ').split('|')[-1].replace('Gross:',' '))
gross_1_to_50  
```

```python
print(len(gross_1_to_50))
```

```python
df1['Votes'] = Votes_1_to_50
df1
```

```python
df1['Gross'] = gross_1_to_50
df1
```
