### Top 10 ODI bowlers along with the records of their team and rating from icc-cricket.com

```python
import numpy as np
import pandas as pd
import requests as rq

from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/bowling')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
odi_bol_table = Data.table
odi_bol_table
```

```python
odi_bol_data = Data.find_all('table',class_='table rankings-table')
odi_bol_data
```

```python
len(odi_bol_data)
```

```python
col = []
for i in odi_bol_table.find_all('th'):
    col.append(i.text)
```

```python
col
```

```python
Date = Data.find('div',class_='rankings-block__last-updated').text
Date
```

```python
odi_bol_list = odi_bol_table.find_all('td')
odi_bol_list
```

```python
odi_bol_list_data = []
start = 0
end = 5
for i in range(20):
    Position,Player,Team,Rating,Career_Best_Rating = odi_bol_list[start:end]
    odi_bol_list_data.append([Position.text.replace('\n',''),Player.text.replace('\n',''),Team.text.replace('\n',''),Rating.text.replace('\n',''),Career_Best_Rating.text.replace('\n','')])
    start +=5
    end +=5
```

```python
odi_bol_list_data
```

```python
df = pd.DataFrame(odi_bol_list_data,columns=col)
df
```

```python
df = df[:10]
df
```

```python
df['Pos'] = df['Pos'].str.strip()
df
```

```python
df['Pos'] = df['Pos'].str.partition(' ')[0]
df
```

```python
df.info()
```

```python
df = df.astype({'Pos':int,'Rating':int})
df
```

```python
df.info()
```

```python
df.rename(columns={'Pos':'Rank'},inplace = True)
```

```python
df1 = df.drop(columns='Career Best Rating')
```

```python
df = df.style.set_caption(Date)
df
```

```python
df.to_excel('4(c). Top 10 ODI bowlers with Career Best Rating.xls')
```
