### Top 10 women’s ODI players along with the records of their team and rating from icc-cricket.com

```python
import numpy as np
import pandas as pd
import requests as rq
import warnings

from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.icc-cricket.com/rankings/womens/player-rankings/odi/batting')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
odi_w_table = Data.table
odi_w_table
```

```python
odi_w_table_data = Data.find_all('table',class_='table rankings-table')
odi_w_table_data
```

```python
len(odi_w_table_data)
```

```python
col = []
for i in Data.find_all("th"):
    col.append(i.text)
```

```python
col
```

```python
odi_w_data = Data.find_all("td")
odi_w_data
```

```python
odi_w_data_list = []
start = 0
end = 5
for i in range(20):
    Position,Player,Team,Rating,Career_Best_Rating = odi_w_data[start:end]
    odi_w_data_list.append([Position.text.replace('\n',''),Player.text.replace('\n',''),Team.text.replace('\n',''),Rating.text.replace('\n',''),Career_Best_Rating.text.replace('\n','')])
    start +=5
    end +=5
```

```python
odi_w_data_list
```

```python
Date = Data.find('div',class_='rankings-block__last-updated').text
Date
```

```python
df = pd.DataFrame(odi_w_data_list,columns=col)
```

```python
df = df[:10]
df
```

```python
df['Pos'] = df['Pos'].str.strip()
df
```

```python
df['Pos'] = df['Pos'].str.partition(' ')[0]
df
```

```python
df.info()
```

```python
df = df.astype({'Pos':int,'Rating':int})
df.info()
```

```python
df
```

```python
df1 = df.drop(columns='Career Best Rating')
df1
```

```python
df.rename(columns={'Pos':'Rank'},inplace=True)
```

```python
df.style.set_caption(Date)
```

```python
df1.rename(columns={'Pos':'Rank'},inplace=True)
```

```python
df1.style.set_caption(Date)
```

```python
df.to_csv('5(b). Top 10 women’s ODI players with Career Best Rating.csv')
```
