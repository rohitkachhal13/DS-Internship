### Top 10 ODI Batsmen in men along with the records of their team and rating icc-cricket.com

```python
import numpy as np
import pandas as pd
import requests as rq
import warnings

from bs4 import BeautifulSoup
warnings.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.icc-cricket.com/rankings/mens/player-rankings/odi/batting')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
odi_table = Data.table
odi_table
```

```python
all_table = Data.find_all('table',class_='table rankings-table')
all_table
```

```python
len(all_table)
```

```python
col = []
for i in odi_table.find_all("th"):
    col.append(i.text)
```

```python
col
```

```python
table_data = odi_table.find_all("td")
table_data
```

```python
table_list = []
start = 0
end = 5

for i in range(20):
    Pos,Player,Team,Rating,Career_Best_Rating = table_data[start:end]
    table_list.append([Pos.text.replace('\n',''),Player.text.replace('\n',''), Team.text.replace('\n',''),Rating.text.replace('\n',''),Career_Best_Rating.text.replace('\n','')])
    start += 5
    end +=5
```

```python
table_list
```

```python
df = pd.DataFrame(table_list,columns=col)
```

```python
df
```

```python
df.rename(columns={'Pos':'Rank'},inplace=True)
```

```python
df
```

```python
df['Rank'] = df['Rank'].str.strip()
df
```

```python
df['Rank'] = df['Rank'].str.partition(' ')[0]
df
```

```python
df = df[0:10]
df
```

```python
df.info()
```

```python
df = df.astype({'Rank':int,'Rating':int})
df.info()
```

```python
df
```

```python
df.to_csv('4(b).Top 10 ODI Batsmen in men with career best rating.csv')
```
