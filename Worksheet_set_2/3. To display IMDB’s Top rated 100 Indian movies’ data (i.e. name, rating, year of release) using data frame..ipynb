### To display IMDB’s Top rated 100 Indian movies’ data (i.e. name, rating, year of release) using data frame.

```python
import pandas as pd
import requests as rq
from bs4 import BeautifulSoup
import warnings
warnings.filterwarnings('ignore')
```

```python
url = rq.get(r'https://www.imdb.com/list/ls051594496/?sort=user_rating,desc&st_dt=&mode=detail')
url
```

```python
Data = BeautifulSoup(url.content)
Data
```

```python
movies = []
for i in Data.find_all('h3',class_='lister-item-header'):
    movies.append(i.text)
movies    
```

```python
movies = []
for i in Data.find_all('h3',class_='lister-item-header'):
    movies.append(i.text.replace('\n',' '))
movies 
```

```python
df = pd.DataFrame({'Movies':movies})
df
```

```python
df['Rank'] = df['Movies'].str.split('.').str.slice(start=0,stop=1).str.join(' ')
df
```

```python
df['Name Of Movie'] = df['Movies'].str.partition('.')[2].str.partition('(2')[0].str.partition('(1')[0]
df
```

```python
df['Releasing Year'] = df['Movies'].str.split('(',n=1).str.get(1).str.replace(')',' ').str.replace('I',' ').str.replace('(',' ')
df
```

```python
length = []
for i in Data.find_all('span',class_='runtime'):
    length.append(i.text)
length    
```

```python
df['Duration(in minutes)'] = length
df
```

```python
df['Duration(in minutes)'] = df['Duration(in minutes)'].str.replace('min','')
```

```python
type_of_movie = []
for i in Data.find_all('span',class_='genre'):
    type_of_movie.append(i.text.replace('\n',' '))
type_of_movie    
```

```python
df['Type Of Movie'] = type_of_movie
df
```

```python
rating = []
for i in Data.find_all('div',class_='ipl-rating-star small'):
    rating.append(i.text.replace('\n',' '))
rating    
```

```python
df['Rating'] = rating
df
```

```python
alls = []
for i in Data.find_all('div',class_='lister-item-content'):
    alls.append(i.text.replace('\n',' '))
alls    
```

```python
df1 = pd.DataFrame({'AX':alls})
df1
```

```python
df['Director'] = df1['AX'].str.partition('Director:')[2].str.partition('|')[0]
df
```

```python
df['Stars'] = df1['AX'].str.partition('Stars:')[2].str.partition('Votes:')[0]
df
```

```python
df['Votes'] = df1['AX'].str.partition('Votes:')[2].str.partition('|')[0]
df
```

```python
df['Gross'] = df1['AX'].str.partition('Gross:')[2]
df
```

```python
df = df.drop(columns='Movies')
df
```

```python
df.info()
```

```python
df['Votes'] = df['Votes'].str.replace(',','')
df
```

```python
df['Releasing Year'] = df['Releasing Year'].str.strip()
```

```python
df['Releasing Year'].value_counts()
```

```python
df['Releasing Year'] = df['Releasing Year'].replace('1986 TV Movie','1986')
```

```python
df['Releasing Year'].value_counts()
```

```python
df = df.astype({'Rank':int,'Releasing Year':int,'Duration(in minutes)':int,'Rating':float,'Votes':int})
df
```

```python
df.to_csv('3. To display IMDB’s Top rated 100 Indian movies’.csv')
```
